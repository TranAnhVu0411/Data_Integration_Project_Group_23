Ta sẽ lấy schema chung cho bảng movie là các trường trong movie data warehouse gồm: 'movie_title', 'movie_info', 'genres', 'directors', 'actors', 'release_date', 'production_company'

Ta sẽ lấy schema chung cho bảng rating là các trường trong movie data warehouse gồm: 'movie_title', 'avg_rating' (rating trung bình), 'rating_count' (số lượt rating), 'link'

```python
# specifying the mappings
data_netflix_movie_mapping = {'Title': 'movie_title',
                              'Summary': 'movie_info',
                              'Genre': 'genres',
                              'Director': 'directors',
                              'Actors': 'actors',
                              'Release Date': 'release_date',
                              'Production House': 'production_company'
                            }
data_imdb_movie_mapping = {'original_title': 'movie_title',
                           'overview': 'movie_info',
                           'genres': 'genres',
                           'director': 'directors',
                           'cast': 'actors',
                           'release_date': 'release_date',
                           'production_companies': 'production_company'
                           }
data_netflix_rating_mapping = {'Title': 'movie_title',
                               'IMDb Score': 'avg_rating',
                               'IMDb Votes': 'rating_count',
                               'Netflix Link': 'link'
                              }
data_imdb_rating_mapping = {'original_title': 'movie_title',
                            'vote_average': 'avg_rating',
                            'vote_count': 'rating_count',
                            'homepage': 'link'
                          }
            
```

```python
# Loại bỏ các trường không cần thiết
keep_field_netflix_movie = list(data_netflix_movie_mapping.keys())
keep_field_imdb_movie = list(data_imdb_movie_mapping.keys())
keep_field_netflix_rating = list(data_netflix_rating_mapping.keys())
keep_field_imdb_rating= list(data_imdb_rating_mapping.keys())

df_netflix_movie_train = df_netflix[keep_field_netflix_movie]
df_imdb_movie_train = df_imdb[keep_field_imdb_movie]
df_netflix_rating_train = df_netflix[keep_field_netflix_rating]
df_imdb_rating_train = df_imdb[keep_field_imdb_rating]
```

```python
# Change type of all columns to string
def create_convert_dict(df):
    convert_dict={}
    for i in df.columns:
        convert_dict[i]=str
    return convert_dict
df_netflix_movie_train = df_netflix_movie_train.astype(create_convert_dict(df_netflix_movie_train))
df_imdb_movie_train = df_imdb_movie_train.astype(create_convert_dict(df_imdb_movie_train))
df_netflix_rating_train = df_netflix_rating_train.astype(create_convert_dict(df_netflix_rating_train))
df_imdb_rating_train = df_imdb_rating_train.astype(create_convert_dict(df_imdb_rating_train))
```

## Traing Phase

Do pandas của flexmatcher không tương thích với pandas hiện tại, cho nên ta phải config file charDistClassifier

url: /venv/lib/python3.9/site-packages/flexmatcher/classify/charDistClassifier.py

- Thay ix() thành iloc()
- Thay as_matrix() thành values

```python
movie_schema_list = [df_netflix_movie_train, df_imdb_movie_train]
movie_mapping_list = [data_netflix_movie_mapping, data_imdb_movie_mapping]
movie_fm = flexmatcher.FlexMatcher(movie_schema_list, movie_mapping_list, sample_size=100)
movie_fm.train()
```

```python
rating_schema_list = [df_netflix_rating_train, df_imdb_rating_train]
rating_mapping_list = [data_netflix_rating_mapping, data_imdb_rating_mapping]
rating_fm = flexmatcher.FlexMatcher(rating_schema_list, rating_mapping_list, sample_size=100)
rating_fm.train()
```

# Schema Matching

```python
# Movie Info
df_metacritic_movie = pd.read_csv("/Users/trananhvu/Documents/Tichhopdulieu/Data_Integration_Group23/Data/metacritic/movieDatasetClean.csv")
df_movielen_movie = pd.read_csv("/Users/trananhvu/Documents/Tichhopdulieu/Data_Integration_Group23/Data/movielen/movies.csv")
df_tmdb_movie = pd.read_csv("/Users/trananhvu/Documents/Tichhopdulieu/Data_Integration_Group23/Data/tmdb/tmdb.csv")
```

```python
# Change types of all columns to string
df_metacritic_movie = df_metacritic_movie.astype(create_convert_dict(df_metacritic_movie))
df_movielen_movie = df_movielen_movie.astype(create_convert_dict(df_movielen_movie))
df_tmdb_movie = df_tmdb_movie.astype(create_convert_dict(df_tmdb_movie))
```

## Movie Schema Matching

```python
predicted_movie_mapping_metacritic = movie_fm.make_prediction(df_metacritic_movie)
predicted_movie_mapping_movielen = movie_fm.make_prediction(df_movielen_movie)
predicted_movie_mapping_tmdb = movie_fm.make_prediction(df_tmdb_movie)
```

```python
# Mapping result
def mapping_result(name, predicted_mapping_result):
    print("Source: "+name)
    result_df = pd.DataFrame(predicted_mapping_result.items(), columns=["Source Schema", "Data Warehouse Schema"])
    return result_df
```

```python
mapping_result("Metacritic", predicted_movie_mapping_metacritic)
```

```python
mapping_result("MovieLen", predicted_movie_mapping_movielen)
```

```python
mapping_result("TMDB", predicted_movie_mapping_tmdb)
```

Kết quả matching của Metacritic và TMDB khá chính xác, trong khi đó của MovieLen bị sai ở phần movieId - release_date

## Rating Schema Matching

```python
predicted_rating_mapping_metacritic = rating_fm.make_prediction(df_metacritic_movie)
predicted_rating_mapping_movielen = rating_fm.make_prediction(df_movielen_movie)
predicted_rating_mapping_tmdb = rating_fm.make_prediction(df_tmdb_movie)
```

```python
mapping_result("Metacritic", predicted_rating_mapping_metacritic)
```

```python
mapping_result("MovieLen", predicted_rating_mapping_movielen)
```

```python
mapping_result("TMDB", predicted_rating_mapping_tmdb)
```

Kết quả matching trong bảng Rating chưa chính xác trong cả 3 trường hợp, cụ thể:

- Metacritic: sai phần year - rating_count và description-movie_title
- MovieLen: sai hết
- TMDB: chỉ chính xác phần vote_average-avg_rating, còn lại sai hết
