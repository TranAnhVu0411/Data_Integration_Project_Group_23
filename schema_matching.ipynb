Ta thấy dữ liệu training của từng nguồn khá cân bằng

Ta sẽ lấy schema chung cho bảng movie là các trường trong movie data warehouse gồm: 'movie_title', 'movie_info', 'genres', 'directors', 'actors', 'release_date', 'runtime', 'content_rating', 'image'

Ta sẽ lấy schema chung cho bảng rating là các trường trong movie data warehouse gồm: 'avg_rating' (rating trung bình), 'rating_count' (số lượt rating), 'link'

```python
# specifying the mappings
data_netflix_movie_mapping = {'Title': 'movie_title',
                              'Summary': 'movie_info',
                              'Genre': 'genres',
                              'Director': 'directors',
                              'Actors': 'actors',
                              'Release Date': 'release_date',
                              'Runtime': 'runtime',
                              'View Rating': 'content_rating',
                              'Image': 'image'
                            }
data_imdb_movie_mapping = {'original_title': 'movie_title',
                           'overview': 'movie_info',
                           'genres': 'genres',
                           'director': 'directors',
                           'cast': 'actors',
                           'release_date': 'release_date',
                           'runtime': 'runtime'
                           }
data_warehouse_movie_mapping = {'movie_title': 'movie_title',
                                'movie_info': 'movie_info',
                                'genres': 'genres',
                                'directors': 'directors',
                                'actors': 'actors',
                                'release_date': 'release_date',
                                'runtime': 'runtime',
                                'content_rating': 'content_rating'
                                }
data_netflix_rating_mapping = {'IMDb Score': 'avg_rating',
                               'IMDb Votes': 'rating_count',
                               'Netflix Link': 'link'
                              }
data_imdb_rating_mapping = {'vote_average': 'avg_rating',
                            'vote_count': 'rating_count',
                            'homepage': 'link'
                          }
data_warehouse_rating_mapping = {'audience_rating': 'avg_rating',
                                 'audience_count': 'rating_count',
                                 'link': 'link'
                                }
            
```

```python
# Loại bỏ các trường không cần thiết
keep_field_netflix_movie = list(data_netflix_movie_mapping.keys())
keep_field_imdb_movie = list(data_imdb_movie_mapping.keys())
keep_field_warehouse_movie = list(data_warehouse_movie_mapping.keys())
keep_field_netflix_rating = list(data_netflix_rating_mapping.keys())
keep_field_imdb_rating= list(data_imdb_rating_mapping.keys())
keep_field_warehouse_rating = list(data_warehouse_rating_mapping.keys())

df_netflix_movie_train = df_netflix[keep_field_netflix_movie]
df_imdb_movie_train = df_imdb[keep_field_imdb_movie]
df_warehouse_movie_train = df_warehouse_movie[keep_field_warehouse_movie]
df_netflix_rating_train = df_netflix[keep_field_netflix_rating]
df_imdb_rating_train = df_imdb[keep_field_imdb_rating]
df_warehouse_rating_train = df_warehouse_rating[keep_field_warehouse_rating]
```

```python
# Change type of all columns to string
def create_convert_dict(df):
    convert_dict={}
    for i in df.columns:
        convert_dict[i]=str
    return convert_dict
df_netflix_movie_train = df_netflix_movie_train.astype(create_convert_dict(df_netflix_movie_train))
df_imdb_movie_train = df_imdb_movie_train.astype(create_convert_dict(df_imdb_movie_train))
df_warehouse_movie_train = df_warehouse_movie_train.astype(create_convert_dict(df_warehouse_movie_train))
df_netflix_rating_train = df_netflix_rating_train.astype(create_convert_dict(df_netflix_rating_train))
df_imdb_rating_train = df_imdb_rating_train.astype(create_convert_dict(df_imdb_rating_train))
df_warehouse_rating_train = df_warehouse_rating_train.astype(create_convert_dict(df_warehouse_rating_train))
```

Ta sẽ lấy schema chung cho bảng review là các trường trong review data warehouse gồm: 'movie_id', 'review_content', 'review_score', 'review_date'

```python
review_data_mapping = {'Movie': 'movie_id',
                       'Review': 'review_content',
                       'Rating': 'review_score'
                      }
review_data_warehouse_mapping = {'movie_id': 'movie_id',
                                 'review_content': 'review_content',
                                 'review_score': 'review_score',
                                 'review_date': 'review_date'
                                }
```

```python
# Loại bỏ các trường không cần thiết
keep_field_review = list(review_data_mapping.keys())
keep_field_review_warehouse = list(review_data_warehouse_mapping.keys())

df_review_train = df_review[keep_field_review]
df_review_warehouse_train = df_review_warehouse[keep_field_review_warehouse]
```

```python
# Change type of all columns to string
df_review_train = df_review_train.astype(create_convert_dict(df_review_train))
df_review_warehouse_train = df_review_warehouse_train.astype(create_convert_dict(df_review_warehouse_train))
```

## Traing Phase

Do pandas của flexmatcher không tương thích với pandas hiện tại, cho nên ta phải config file charDistClassifier

url: /venv/lib/python3.9/site-packages/flexmatcher/classify/charDistClassifier.py

- Thay ix() thành iloc()
- Thay as_matrix() thành values

### Movie and Rating Schema Matching

```python
movie_schema_list = [df_netflix_movie_train, df_warehouse_movie_train, df_imdb_movie_train]
movie_mapping_list = [data_netflix_movie_mapping ,data_warehouse_movie_mapping, data_imdb_movie_mapping]
movie_fm = flexmatcher.FlexMatcher(movie_schema_list, movie_mapping_list, sample_size=300)
movie_fm.train()
```

```python
rating_schema_list = [df_netflix_rating_train, df_warehouse_rating_train, df_imdb_rating_train]
rating_mapping_list = [data_netflix_rating_mapping, data_warehouse_rating_mapping, data_imdb_rating_mapping]
rating_fm = flexmatcher.FlexMatcher(rating_schema_list, rating_mapping_list, sample_size=300)
rating_fm.train()
```

### Review Schema Matching

```python
review_schema_list = [df_review_train, df_review_warehouse_train]
review_mapping_list = [review_data_mapping, review_data_warehouse_mapping]
review_fm = flexmatcher.FlexMatcher(review_schema_list, review_mapping_list, sample_size=300)
review_fm.train()
```

# Schema Matching

```python
# Movie Info
df_metacritic_movie = pd.read_csv(os.path.join(folder.metacritic_dir, "movieDatasetClean.csv"))
df_movielen_movie = pd.read_csv(os.path.join(folder.movielen_dir, "movies.csv"))
df_tmdb_movie = pd.read_csv(os.path.join(folder.tmdb_dir,"tmdb.csv"))

# Review Info
df_metacritic_review = pd.read_csv(os.path.join(folder.metacritic_dir, "reviewDatasetClean.csv"), 
                                   lineterminator='\n')
df_movielen_rating = pd.read_csv(os.path.join(folder.movielen_dir,"ratings.csv"))
```

```python
# Change types of all columns to string
df_metacritic_movie = df_metacritic_movie.astype(create_convert_dict(df_metacritic_movie))
df_movielen_movie = df_movielen_movie.astype(create_convert_dict(df_movielen_movie))
df_tmdb_movie = df_tmdb_movie.astype(create_convert_dict(df_tmdb_movie))

df_metacritic_review = df_metacritic_review.astype(create_convert_dict(df_metacritic_review))
df_movielen_rating = df_movielen_rating.astype(create_convert_dict(df_movielen_rating))
```

## Movie Schema Matching

```python
predicted_movie_mapping_metacritic = movie_fm.make_prediction(df_metacritic_movie)
predicted_movie_mapping_movielen = movie_fm.make_prediction(df_movielen_movie)
predicted_movie_mapping_tmdb = movie_fm.make_prediction(df_tmdb_movie)
```

```python
# Mapping result
def mapping_result(name, predicted_mapping_result):
    print("Source: "+name)
    result_df = pd.DataFrame(predicted_mapping_result.items(), columns=["Source Schema", "Data Warehouse Schema"])
    return result_df
```

```python
mapping_result("Metacritic", predicted_movie_mapping_metacritic)
```

```python
mapping_result("MovieLen", predicted_movie_mapping_movielen)
```

```python
mapping_result("TMDB", predicted_movie_mapping_tmdb)
```

Kết quả matching của cả 3 nguồn gần chính xác, chỉ mắc một số sai sót như:
- Metacritic bị sai ở phần rating - content_rating, url - movie_title
- MovieLen bị sai ở phần movieId - runtime
- TMDB chỉ đúng title - movie_title, overview - movie_info, release_date - release_date, genres - genres

## Rating Schema Matching

```python
predicted_rating_mapping_metacritic = rating_fm.make_prediction(df_metacritic_movie)
predicted_rating_mapping_movielen = rating_fm.make_prediction(df_movielen_movie)
predicted_rating_mapping_tmdb = rating_fm.make_prediction(df_tmdb_movie)
```

```python
mapping_result("Metacritic", predicted_rating_mapping_metacritic)
```

```python
mapping_result("MovieLen", predicted_rating_mapping_movielen)
```

```python
mapping_result("TMDB", predicted_rating_mapping_tmdb)
```

Kết quả matching trong bảng Rating chưa chính xác trong cả 3 trường hợp, cụ thể:

- Metacritic: sai phần year - rating_count
- MovieLen: sai hết
- TMDB: chỉ chính xác phần vote_average-avg_rating, còn lại sai hết

## Review Schema Matching

```python
predicted_review_mapping_metacritic = review_fm.make_prediction(df_metacritic_review)
predicted_rating_mapping_movielen = review_fm.make_prediction(df_movielen_rating)
```

```python
mapping_result("Metacritic", predicted_review_mapping_metacritic)
```

```python
mapping_result("MovieLen", predicted_rating_mapping_movielen)
```

Kết quả matching trong bảng Rating chưa chính xác trong cả 2 trường hợp, cụ thể:

- Metacritic: chỉ đúng review - review_content và title - movie_id
- MovieLen: chỉ đúng rating - review_score
